```python
import tensorflow as tf
import numpy as np
import matplotlib as plt

INPUT_SHAPE = ()
LeakyReLU = tf.keras.layers.LeakyReLU(alpha=0.1)

#create model
model = tf.keras.Sequential()

#input layer
model.add(tf.keras.layers.Conv3D(input_shape=(INPUT_SHAPE), activation=LeakyReLU))

#hidden layers
#model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
#model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(16000, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(4000, activation=tf.nn.relu))

#output layer
model.add(tf.keras.layers.Dense(1, activation=tf.nn.relu))

#compile model with optimizer, loss, and metrics
model.compile(optimizer='adam',
              loss='mean_absolute_error',
              metrics=['accuracy'])
```

```python
#load training data into model
images = np.load("traindata-80x80x30.npy", allow_pickle=True)

#parse training data
x_train = []
y_train = []
b = []
for n in range(len(images)):
    x_train.append(images[n, 0])
    y_train.append(images[n, 1])
    b.append(images[n, 2])
X = np.array(x_train)
Y = np.array(y_train)
```

```python
#train model
model.fit(x=X,
         y=Y,
         epochs=2,
         shuffle=True)
```
